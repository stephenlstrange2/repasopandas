#sacar balance promedio de los clientes

```python
cuentas = pd.read_csv('Ejercicio.csv')


sum(cuentas['CURRENT_BALANCE'].notnull()) # aqui sacamos la suma en la que no son nulos
cuentas[cuentas['CURRENT_BALANCE'].notnull()]['CURRENT_BALANCE'].min() #esta es el minimo 
#vamos a sustitur los datos de current balance donde vemos nan por ceros

cuentas['CURRENT_BALANCE']=cuentas['CURRENT_BALANCE'].fillna(0)

query = cuentas['CURRENT_BALANCE'] # aqui estamos sacando los balances
query.mean()   #aqui calculamos el balance promedio
query.max()
print("el promedio de current balance es "+str(round(query.mean(),2))) ##aqui redondeamos el resultado a dos decimales y lo imprimimos
```

Sacar los que tienen account id pares

```python
#vamos a hacerlo con dos metodos

#el primer metodo es directamente en el query
cuentas = pd.read_csv('Ejercicio.csv') #leemos el archivo
pares = cuentas[cuentas['ACCOUNT_ID']%2 == 0] #hacemos modulo 2 a las cuentas y queriamos 
set(pares['ACCOUNT_ID']) #separamos para no contar los repetidos


#segundo metodo es a traves de una lambda

cuentas[cuentas['ACCOUNT_ID'].apply(lambda x : x%2 == 0)] #es una funcion expresada de manera regular en una sola linea

```

#hacemos una funcion que sea el cuadrado del account id

```python
cuentas = pd.read_csv('Ejercicio.csv') #leemos el archivo
def cuadrado(x):  #definimos una funcion que eleva al cuadrado el argumento
    return x**2 
cuentas['cuadrado']=cuentas['ACCOUNT_ID'].apply(cuadrado) 
# en apply mandamos a una funcion, en este caso como cuentas_account id regresa un numero es implicita la llamada
cuentas
```

#clasificar los pares e impares en el numero de cuenta

```python
cuentas = pd.read_csv('Ejercicio.csv') #leemos el archivo
 #se crea una funcion en la que se saca el modulo 2 de el numero de serie, y en caso true se agrega en la nueva columna modulo Par y en caso contrario impar
cuentas['modulo']= cuentas['ACCOUNT_ID'].apply(lambda x :"PAR" if x%2 == 0 else "IMPAR")
cuentas
```
